```python
import numpy as np
import matplotlib.pyplot as plt
```

**Dataset Loading**

```python
# Dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# Targets
y_and = np.array([0,0,0,1])
y_or  = np.array([0,1,1,1])

# Sigmoid
def sigmoid(z):
    return 1/(1+np.exp(-z))

# Training function
def train(y, lr):

    w = np.array([0.1, 0.2])
    b = 0.3
    epochs = 75
    losses = []

    for epoch in range(epochs):
        total_loss = 0

        for i in range(len(X)):
            z = np.dot(X[i], w) + b
            y_pred = sigmoid(z)

            # Binary Cross Entropy Loss
            loss = -(y[i]*np.log(y_pred) + (1-y[i])*np.log(1-y_pred))
            total_loss += loss

            # Backprop
            dz = y_pred - y[i]
            dw = dz * X[i]
            db = dz

            # Update
            w -= lr * dw
            b -= lr * db

        losses.append(total_loss/len(X))

    return w, b, losses

# Learning rates
lr_list = [0.1, 0.7, 0.5,0.9]
```

**Implementation of Loss / Epochs for AND gate**

```python
# ---------- AND Gate ----------
print("\n===== AND GATE =====")
plt.figure()

for lr in lr_list:
    w, b, losses = train(y_and, lr)
    plt.plot(losses, label=f"LR={lr}")

    # Numerical outputs
    print(f"\nLearning Rate: {lr}")
    print("Final Weights:", w)
    print("Final Bias:", b)

    print("Predicted Outputs:")
    for i in range(len(X)):
        pred = sigmoid(np.dot(X[i], w) + b)
        print(f"Input {X[i]} -> {pred:.4f}")

    print("Final Loss:", losses[-1])
plt.title("AND Gate: Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

**Implementation of Loss / Epochs for OR gate**

```python
# ---------- OR Gate ----------
print("\n===== OR GATE =====")
plt.figure()

for lr in lr_list:
    w, b, losses = train(y_or, lr)
    plt.plot(losses, label=f"LR={lr}")

    # Numerical outputs
    print(f"\nLearning Rate: {lr}")
    print("Final Weights:", w)
    print("Final Bias:", b)

    print("Predicted Outputs:")
    for i in range(len(X)):
        pred = sigmoid(np.dot(X[i], w) + b)
        print(f"Input {X[i]} -> {pred:.4f}")

    print("Final Loss:", losses[-1])

plt.title("OR Gate: Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
