```python
# Performing Multi Layer Perceptron ( 3 neurons ) using Step Function
```

```python
def step(x):
    if x>=0:
        return 1
    else:
        return 0
```

```python
def xor_mlp(x1,x2):
    print(f"INPUTS : x1 = {x1}, x2 = {x2}")

    z1 = (x1*1) + (x2*1) - 0.5
    h1 = step(z1)
    print("\nNeuron 1 (OR Gate)")
    print(f"Weighted Sum (z1) = {x1}*1 + {x2}*1 - 0.5 = {z1}")
    print(f"Output after Step Activation (h1) = {h1}")

    z2 = (x1 * -1 ) + ( x2* -1 ) + 1.5
    h2 = step(z2)
    print("\nNeuron 2 (NAND Gate)")
    print(f"Weighted Sum (z2) = {x1}*(-1) + {x2}*(-1) + 1.5 = {z2}")
    print(f"Output after Step Activation (h2) = {h2}")
    
    z3  = (h1 * 1) + (h2 *1 ) - 1.5
    y = step(z3)
    print("\nNeuron 3 (AND Gate - Final XOR Output)")
    print(f"Weighted Sum (z3) = {h1}*1 + {h2}*1 - 1.5 = {z3}")
    print(f"Final Output after Step Activation (XOR) = {y}\n")

    return y
```

```python
print ( "XOR gate using Multi Layer Perceptron\n")
inputs = [(0,0), (0,1), (1,0), (1,1)]
for x1,x2 in inputs:
    xor_mlp(x1,x2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def step(x):
    return np.where(x >= 0, 1, 0)

# Input range
x = np.linspace(-10, 10, 100)

# Output
y = step(x)

# Plot
plt.figure()
plt.plot(x, y)
plt.title("Step Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
# Performing Multi Layer Perceptron ( 3 neurons ) using Sigmoid Function
```

```python
import math
def sigmoid(x):
    return 1/(1 + math.exp(-x))
```

```python
def xor_mlp_sigmoid(x1,x2):
    print(f"Inputs: x1 = {x1}, x2 = {x2}")

    z1 = (x1*1) + (x2*1) -0.5
    h1 = sigmoid(z1)
    print("\nNeuron 1 (OR Gate)")
    print(f"Weighted Sum (z1) = {z1}")
    print(f"Sigmoid Output (h1) = {h1}")

    z2 = (x1 * -1) + (x2 * -1) + 1.5
    h2 = sigmoid(z2)
    
    print("\nNeuron 2 (NAND Gate)")
    print(f"Weighted Sum (z2) = {z2}")
    print(f"Sigmoid Output (h2) = {h2}")

    z3 = (h1 * 1) + (h2 * 1) - 1.5
    y_raw = sigmoid(z3)
    
    # Thresholding at 0.5
    y = 1 if y_raw >= 0.5 else 0
    
    print("\nNeuron 3 (AND Gate - Final Output)")
    print(f"Weighted Sum (z3) = {z3}")
    print(f"Sigmoid Output (Before Threshold) = {y_raw}")
    print(f"Final Binary Output (After Threshold) = {y}")
    
    return y
```

```python
print("XOR Gate using Multi-Layer Perceptron (Sigmoid Activation)\n")

inputs = [(0,0), (0,1), (1,0), (1,1)]

for x1, x2 in inputs:
    xor_mlp_sigmoid(x1, x2)
```
