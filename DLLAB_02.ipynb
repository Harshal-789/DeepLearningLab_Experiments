**Perceptron Learning Algorithm for implementing AND gate**

```python
import numpy as np
# for AND gate (Input/Output)
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])
epochs = 4
w,b = np.zeros(2),0

for _ in range(epochs):
  for x,y in zip(x_list,y_list):
    z = np.dot(x,w) + b
    ### condition area for y_pred
    y_pred = 1 if z >= 0 else 0
    E = y - y_pred
    print("error : ",E,"<- at epoch",_)
    w += E*x
    b += E
    print(w,b)
```

**Change the random weights and biases and observe the convergence of the network.**

```python
import numpy as np

# AND gate dataset
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])

epochs = 10

np.random.seed(1)
w = np.random.randn(2)
b = np.random.randn()

print("Initial weights:", w)
print("Initial bias:", b)
print("-" * 40)

for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    total_error = 0

    for x, y in zip(x_list, y_list):
        z = np.dot(x, w) + b
        y_pred = 1 if z >= 0 else 0
        E = y - y_pred
        total_error += abs(E)

        w += E * x
        b += E

        print(f"x={x}, y={y}, y_pred={y_pred}, error={E}")
        print("Updated w:", w, "Updated b:", b)

    print("Total error in this epoch:", total_error)
```
